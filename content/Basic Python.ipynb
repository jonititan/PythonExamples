```python
import sympy as sp
from sympy import init_printing
init_printing()
from sympy.physics.units import kilogram, meter, second

# Define symbolic variables (positive for physical sense)
rho, V, L, mu = sp.symbols(r'\rho V L \mu', positive=True)

# Attach full SI units
rho_si = (rho * kilogram) / meter**3          # density
V_si   = (V * meter) / second                 # velocity
L_si   = L * meter                          # characteristic length (e.g. chord)
mu_si  = (mu * kilogram )/ (meter * second)   # dynamic viscosity

# Reynolds number expression (units are multiplied in)
Re_expr = (rho_si * V_si * L_si) / mu_si
```

```python
Re_expr
```

```python
# Example: compute numerical Re for a wing (sea-level standard day)
subs_dict = {rho: 1.225, V: 60, L: 1.2, mu: 1.7894e-5}
Re_value = Re_expr.subs(subs_dict).doit().evalf()
print(f"Reynolds number = {Re_value:.2e}")   # → ~4.93e6
```
